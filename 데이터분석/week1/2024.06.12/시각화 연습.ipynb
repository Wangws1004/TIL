```python
# 4. 데이터 시각화 
# 1000만원 이하의 월평균 근로소득의 분포를 히스토그램으로 그리고 
# 평균과 중앙값을 선으로 표시하세요

mean_1000 = df[df['월평균 근로소득_만원'] <= 1000]
mean_1000.head()
```

```python
mean_val = mean_1000['월평균 근로소득_만원'].mean()
median_val = mean_1000['월평균 근로소득_만원'].median()

print(mean_val, median_val)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(mean_1000['월평균 근로소득_만원'], bins=100, edgecolor='black')

plt.axvline(mean_val, color='green', linestyle='dashed', linewidth=1, label=f'평균: {mean_val:.2f}')
plt.axvline(median_val, color='red', linestyle='dashed', linewidth=1, label=f'중앙값: {median_val:.2f}')

plt.title('1000만원 이하의 월평균 근로소득 분포')
plt.xlabel('월평균 근로소득_만원')
plt.ylabel('Count')
plt.legend()

plt.show()
```

```python
# 5. 전공계열에 따른 월평균 근로소득 분포
# 전공계열에 따른 월평균 근로소득의 분포를 바이올린 플롯으로 그리고, 
# 이직여부를 구분하여 그리십시오.(sns.violinplot)

plt.figure(figsize=(12, 6))
sns.violinplot(x='전공계열', y='월평균 근로소득_만원', hue='이직여부', data=df)

plt.title('전공계열에 따른 월평균 근로소득 분포')
plt.xlabel('전공계열')
plt.ylabel('월평균 근로소득 (만원)')
plt.legend(title='이직여부')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='전공계열', y='월평균 근로소득_만원', hue='이직여부', data=df)
sns.stripplot(x='전공계열', y='월평균 근로소득_만원', data=df, alpha=0)

plt.title('전공계열에 따른 월평균 근로소득 분포')
plt.xlabel('전공계열')
plt.ylabel('월평균 근로소득 (만원)')
plt.legend(title='이직여부')

plt.show()
```

```python
# 6. 주당 정규 근로시간에 따른 월평균 근로소득 분포
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="주당 정규 근로시간", y="월평균 근로소득_만원", alpha=0.8, hue="이직여부")
plt.xlabel('주당 정규 근로시간')
plt.ylabel('월평균 근로소득_만원')
plt.show()
```

```python
# 7. 월평균 근로소득에 영향을 미치는 변수
# 월평균 근로소득_만원', '주당 정규 근로시간', '졸업 평점', '전반적 만족도(직장)'의 쌍별 관계를 
# pairplot을 이용해 시각화하고, 이직여부를 구분하여 그리십시오.
# 월 근로소득 1000만원 이하만 조사한다
# mean_1000 = df[df['월평균 근로소득_만원'] <= 1000]

sns.pairplot(data=mean_1000, vars=['월평균 근로소득_만원', '주당 정규 근로시간', '졸업 평점', '전반적 만족도(직장)'], hue="이직여부")
plt.show()
```

```python
# 8. 전공계열의 분포
# 전공계열의 분포를 막대그래프와 파이차트로 각각 표현하십시오.
count = df['전공계열'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=count.index, y=count.values)
plt.xlabel('전공계열')
plt.ylabel('count')
plt.show()
```

```python
# 9. 이직여부의 분포
# - 이직여부의 분포를 막대그래프와 파이차트로 각각 표현하십시오
val = df['이직여부'].value_counts()

sns.barplot(x=val.index, y=val.values)
plt.show()
```

```python
# autopct -> 퍼센티지 표시
plt.pie(val, labels=['이직x', '이직o'], autopct='%1.1f%%')
plt.show()
```
