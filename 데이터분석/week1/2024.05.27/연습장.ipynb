```python
from bs4 import BeautifulSoup
import requests

# 검색어와 페이지 수 입력 받기
query = input("검색어를 입력 하세요 : ")
pages = int(input("몇 페이지까지 검색하실 건가요? : "))

# 결과를 저장할 리스트 초기화
all_results = []

# 페이지 수만큼 반복
for page in range(1, pages + 1):
    # print(page)
    # 각 페이지의 URL 생성
    hankyung_url = f"https://search.hankyung.com/search/news?query={query}&page={page}"
    # 페이지 요청 및 파싱
    hankyung_raw = requests.get(hankyung_url)
    hankyung_bs = BeautifulSoup(hankyung_raw.text, "html.parser")
    # print(hankyung_bs.text)

    news = hankyung_bs.find_all("div", class_='txt_wrap')

    for i in news:
      title = i.find('em', class_='tit').text
      content = i.find('p', class_='txt').text
      all_results.append([title, content])

# 최종 결과 출력
for result in all_results:
    print(result)
```

```python
import requests
from bs4 import BeautifulSoup

def fetch_page(query, page):
    hankyung_url = f"https://search.hankyung.com/search/news?query={query}&page={page}"
    hankyung_raw = requests.get(hankyung_url)
    hankyung_bs = BeautifulSoup(hankyung_raw.text, "html.parser")
    return hankyung_bs

def parse_news(hankyung_bs):
    results = []
    news = hankyung_bs.find_all("div", class_='txt_wrap')

    for item in news:
        title = item.find('em', class_='tit').text
        content = item.find('p', class_='txt').text
        results.append([title, content])

    return results

def search_news(query, pages):
    all_results = []

    for page in range(1, pages + 1):
        hankyung_bs = fetch_page(query, page)
        results = parse_news(hankyung_bs)
        all_results.extend(results)

    return all_results

def main():
    # 검색어와 페이지 수 입력 받기
    query = input("검색어를 입력하세요: ")
    pages = int(input("몇 페이지까지 검색하실 건가요? "))

    # 뉴스 검색 및 결과 출력
    results = search_news(query, pages)

    for result in results:
        print(f"Title: {result[0]}\nContent: {result[1]}\n")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent

def get_user_agent():
    ua = UserAgent()
    headers = {
        "User-Agent": ua.random
    }
    return headers

def get_gens():
    genre_list = {
        "1": "GN0100",
        "2": "GN0200",
        "3": "GN0300",
        "4": "GN0400",
        "5": "GN0500",
        "6": "GN0600",
        "7": "GN0700",
        "8": "GN0800"
    }

    print("장르:")
    print("1. 발라드")
    print("2. 댄스")
    print("3. 랩/힙합")
    print("4. R&B/Soul")
    print("5. 인디음악")
    print("6. 록/메탈")
    print("7. 트로트")
    print("8. 포크/블루스")

    genre = input("원하는 장르를 입력하세요 (번호 선택): ")

    return genre_list.get(genre, "GN0100")

def get_song_list(genre_code, headers):
    url = f"https://www.melon.com/genre/song_list.htm?gnrCode={genre_code}"
    res = requests.get(url, headers=headers)
    res.raise_for_status()  # 요청 실패 시 예외 발생
    bs = BeautifulSoup(res.text, 'html.parser')
    area = bs.select_one(".service_list_song")
    songs = area.select("div > table > tbody > tr")

    return songs

def get_table(songs):
    song_list = []
    for e in songs:
        rank = e.select_one(".rank").text.strip()
        title = e.select_one(".ellipsis.rank01 > span > a").text.strip()
        singer = e.select_one(".ellipsis.rank02 > a").text.strip()
        album = e.select_one(".ellipsis.rank03 > a").text.strip()

        song_list.append([rank, title, singer, album])

    return song_list

def print_table(song_list):
    for song in song_list:
        print(f"순위: {song[0]}, 제목: {song[1]}, 가수: {song[2]}, 앨범: {song[3]}")

def main():
    headers = get_user_agent()
    genre_code = get_gens()
    songs = get_song_list(genre_code, headers)
    song_list = get_table(songs)
    print_table(song_list)

if __name__ == "__main__":
    main()
```
