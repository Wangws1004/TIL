## 막대 그래프
- 막대 그래프는 범주별로 도수, 횟수, 퍼센트 등의 크기만큼 막대를 그리는 그래프
- 범주별 미세한 차이를 명확하게 전달할 수 있는 그래프

```python
from matplotlib import pyplot as plt
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
x_len = range(len(movies))
# plt 객체 생성 변수에 별도로 할당하지는 않는다. 
# 셀 안에 한번에 모든 내용들 넣어둔다. 
# plt 객체 에서 어떤 그래프를 그릴지 결정하고 code를 작성
# plt에 속성값 들을 지정 해준다.
# plt.show() 셀안의 속성들이 지정된 그래프를 출력
# plt 객체는 cell 단위로 실행 된다. 

plt.bar(x_len, num_oscars)
plt.xticks(x_len, movies)
plt.xlabel("movie name")
plt.ylabel("Num of Oscars")
plt.show()
```

## 히스토그램

- 막대 그래프의 한 종류
- 양적 데이터를 범주로 사용할 때 유용하다.
    - 예를 들어, 무게, 길이, 시간, 점수 등 
    
- 숫자의 범위에 따라 나뉘어진 데이터를 막대 그래프로 보여줄 때 사용한다.
    - 양적 데이터들의 연속성을 반영하기 위해 막대 그래프들 사이의 간극이 없도록 표현한다.
    - 정해진 구간에 해당되는 항목의 개수나 빈도수를 보여주어 값의 분포를 쉽게 관찰
    
__참고__
- 질적 자료
    - 사칙 연산이 불가능
    - 비계측 자료(Nonmetric Data) 혹은 범주형 자료(Categorical Data)
    - 예) 직업, 종교, 성별

- 양적 자료
    - 사칙 연산 가능
    - 계측 자료(Metric Data) 혹은 숫자형 자료(Numerical Data)
    - 예) 몸무게, 용돈, 가격

위 데이터를 히스토그램으로 나타내면 다음과 같다.

* x 축: 점수 구간
* y 축: 학생 수

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0, 10, 90]

histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
```

```python
histogram
```

```python
grade_key = histogram.keys()
```

```python
grade_value = histogram.values()
```

```python
import numpy as np
```

```python
np.array(grade_key)
```

```python
plt.bar(np.array(list(grade_key))+ 5, grade_value, 10, edgecolor = (0, 0, 0))
plt.xticks([10*i for i in range(11)])
plt.show()
```

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends, minutes)
```

```python
plt.scatter(friends, minutes)
for l, f, m in zip(labels, friends, minutes):
    plt.annotate(l, xy=(f, m), textcoords="offset points", xytext=(5, -5))
plt.show()
```
