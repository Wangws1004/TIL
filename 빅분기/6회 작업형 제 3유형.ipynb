```python
# A도시의 교통수단은 버스, 지하철, 택시, 자전거
# 각 교통수단의 실제 선택 비율이 각 40% 30% 20% 10%
# 이를 검증하기 위해 대중교통 이용객들이 선호하는 교통수단에 대해 설문조사를 했다.
# 이를 위해 수립한 귀무가설은 다음과 같다

# H0 : 대중교통 이용객의 실제 교통수단 선택 비율이 예상 비용과 같다.
# H1 : 대중교통 이용객의 실제 교통수단 선택 비율이 예상 비율과 다르다.

# 이에 대한 검정을 진행하시오.
```

```python
# 주어진 설문조사 데이터에서 전체 건수 대비 Subway를 선호하는 이용자의 비율을 구하시오(반올림 소수점 둘째자리까지 계산)

import pandas as pd
df = pd.read_csv('6_3_1_transport_data.csv')
df.head()
```

```python
# subway를 선호하는 이용자의 비율
subway_pref = (df['TransportMode'] == 'subway').mean()
res = round(subway_pref, 2)
print(res)
```

```python
# 예상 이용자 수와 실제 이용자 수가 동일한 비율을 보이는지 적합도를 검정하고 검정 통계량을 소수 둘째 자리로 반올림하시오
```

```python
from scipy.stats import chisquare

observed_counts = df['TransportMode'].value_counts()

observed_counts = observed_counts.reindex(['bus', 'subway', 'taxi', 'bicycle']).fillna(0)
total_response = df.shape[0]
expected_rates = [0.4, 0.3, 0.2, 0.1] # 교통 수단 별 기대비율
expected_counts = [rate * total_response for rate in expected_rates]
```

```python
total_response
```

```python
expected_counts
```

```python
# 적합도 검정 수행
chi2_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)
res = round(chi2_stat, 2)
print(res)
```

```python
# 위의 통계량에 대한 p-value를 구하고, 유의수준 0.05하에서 가설 검정의 결과를 채택, 기각 중 하나로 선택하시오
# p-value값은 소수점 둘째 자리까지 계산한다.

p_value_rounded = round(p_value, 2)

if p_value_rounded > 0.05:
  decision = '귀무 가설 수용: 실제 교통수단 선택 비율이 기대 비율과 유의미하게 다르지 않음'
else:
  decision = '귀무 가설 기각: 실제 교통 수단 선택 비율이 기대 비율과 유의미하게 다름'

print(f'p-값: {p_value_rounded}, 결정: {decision}')
```

```python
# Length의 회귀계수를 구하고 소수 둘째 자리로 반올림
# statsmodels 라이브러리
```

```python
import pandas as pd
import statsmodels.api as sm

df2 = pd.read_csv('6_3_2_fish_weight_data.csv')
df2.head()
```

```python
X = df2[['Length', 'Height', 'Width', 'FinSize']]
Y = df2['Weight']

X = sm.add_constant(X)
```

```python
model = sm.OLS(Y, X).fit()
```

```python
res = round(model.params['Length'], 2)
print(res)
```

```python
# p-value 계산
res = model.pvalues['Height']
print(res)
```

```python
res = 'Y' if model.pvalues['Width'] < 0.05 else 'N'
print(res)
```
