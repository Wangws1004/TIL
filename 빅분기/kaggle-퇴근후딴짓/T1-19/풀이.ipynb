### 주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과
### 2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오.
### 단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함
### (최종값은 소수점 반올림 후 정수 출력)
- 데이터셋 : basic2.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
import pandas as pd
import numpy as np

df = pd.read_csv("basic2.csv")
df.head()
# 1. 2022년 월별 Sales 합계 중 가장 큰 금액과
# 2. 2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값
# 3. 단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

df.head()
```

```python
def event_sales(x):
  if x['Events'] == 1:
    x['Sales2'] = x['Sales'] * 0.8
  else:
    x['Sales2'] = x['Sales']
  return x
```

```python
df = df.apply(lambda x: event_sales(x), axis=1) #1일 경우 row, 0일 경우 컬럼
df.head()
```

```python
df_2022 = df[df['Year'] == 2022].groupby('Month')['Sales2'].sum().max()
df_2022
```

```python
df_2023 = df[df['Year'] == 2023].groupby('Month')['Sales2'].sum().max()
df_2023
```

```python
print(round(abs(df_2022 - df_2023)))
```
