```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df['year'] = df['Date'].dt.year == 2022
df['month'] = df['Date'].dt.month == 5
df
```

```python
sales = df[(df['year'] == True) & (df['month'] == True)].median()[1]
sales
```
