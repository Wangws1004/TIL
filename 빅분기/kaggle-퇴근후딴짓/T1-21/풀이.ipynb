## 나이 구간 나누기
###  basic1 데이터 중 'age'컬럼 이상치를 제거하고, 동일한 개수로 나이 순으로 3그룹으로 나눈 뒤 각 그룹의 중앙값을 더하시오
### (이상치는 음수(0포함), 소수점 값)
- data: basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작


```python
import pandas as pd
import numpy as np

df = pd.read_csv("basic1.csv")
df.head()
```

```python
df['age'].isna().sum()
```

```python
df = df[~(df['age'] <= 0)]
df = df[df['age'] == round(df['age'], 0)]

df.head()
```

```python
pd.qcut(df['age'], q=3)
```

```python
# 기준 확인
pd.qcut(df['age'], q=3)
```

```python
df['range'] = pd.qcut(df['age'], q=3, labels=['group1', 'group2', 'group3'])
df.head()
```

```python
group1 = df[df['range'] == 'group1']['age'].median()
group2 = df[df['range'] == 'group2']['age'].median()
group3 = df[df['range'] == 'group3']['age'].median()

print(group1 + group2 + group3)
```
