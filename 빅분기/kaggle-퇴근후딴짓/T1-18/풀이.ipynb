## 주어진 데이터에서 2022년 5월 주말과 평일의 sales컬럼 평균값 차이를 구하시오 (소수점 둘째자리까지 출력, 반올림)
- 데이터셋 : basic2.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
import pandas as pd
import numpy as np

df = pd.read_csv("basic2.csv")
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.dayofweek

df.head()
```

```python
df['weekend'] = df['dayofweek'] >= 5
df.head()
```

```python
sales_weekend = df[(df['Year'] == 2022) & (df['Month'] == 5) & (df['weekend'] == True)]['Sales'].mean()
sales_weekend
```

```python
sales_week = df[(df['Year'] == 2022) & (df['Month'] == 5) & (df['weekend'] == False)]['Sales'].mean()
sales_week
```

```python
print(round(sales_weekend - sales_week, 2))
```
