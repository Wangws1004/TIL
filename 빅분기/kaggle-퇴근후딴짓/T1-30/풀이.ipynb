# T1-30. 12월 25일 결제 금액(price)은 12월 총 결제금액의 몇 %인가? (정수로 출력)

```python
import pandas as pd
df = pd.read_csv("payment.csv")
df
```

```python
# your code
```

```python
df['date'] = pd.to_datetime(df['date'], format="%Y%m%d")
df.head()
```

```python
d1 = df['date'].dt.month == 12
d2 = df['date'].dt.day == 25

df[d1 & d2]
```

```python
sum_1225 = df[d1 & d2]['price'].sum()
sum_12 = df[d1]['price'].sum()

print(int((sum_1225 / sum_12) * 100))
```
