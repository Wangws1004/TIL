## basic1 데이터에서 f4가 E로 시작하면서 부산에 살고 20대인 사람은 몇 명일까요?

```python
import pandas as pd
df = pd.read_csv("basic1.csv")

df.head()
```

```python
df['mbti'] = df['f4'].str[:1]
df.head()
```

```python
res = df[(df['mbti'] == "E") & (df['city'] == "부산") & (df['age'] >= 20) & (df['age'] < 30)]
len(res)
```
