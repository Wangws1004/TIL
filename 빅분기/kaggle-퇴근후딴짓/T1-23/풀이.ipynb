## f1의 결측치를 채운 후 age 컬럼의 중복 제거 전과 후의 'f1' 중앙값 차이를 구하시오
### - 결측치는 f1의 데이터 중 10번째로 큰 값으로 채움
### - 중복 데이터 발생시 뒤에 나오는 데이터를 삭제함
### - 최종 결과값은 절대값으로 출력

- 데이터셋 : basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작

```python
import pandas as pd
import numpy as np


df = pd.read_csv("basic1.csv")
df.head()
```

```python
df['f1'].isna().sum()
```

```python
sort_df = df.sort_values(ascending=False, by='f1')
sort_df.head(11)
```

```python
f1_10 = sort_df['f1'].iloc[9]
f1_10
```

```python
df_fill = df.fillna(f1_10)
df_fill
```

```python
before = df_fill['f1'].median()
before
```

```python
after = df_fill.drop_duplicates(subset=['age'])
after
```

```python
res = after['f1'].median()
res
```

```python
print(abs(before - res))
```
