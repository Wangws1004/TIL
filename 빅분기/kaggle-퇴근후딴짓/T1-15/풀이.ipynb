주어진 데이터 셋에서 age컬럼 상위 20개의 데이터를 구한 다음 <br>
f1의 결측치를 중앙값으로 채운다. <br>
그리고 f4가 ISFJ와 f5가 20 이상인 <br>
f1의 평균값을 출력하시오!

- 데이터셋 : basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작
- File -> Editor Type -> Script


```python
import pandas as pd

df = pd.read_csv("basic1.csv")

df.head()
```

```python
 age_top20 = df.sort_values('age', ascending=False).head(20)
 age_top20
```

```python
age_top20.info()
```

```python
med = age_top20['f1'].median()
med
```

```python
age_top20['f1'].fillna(med, inplace=True)
age_top20
```

```python
# f4가 ISFJ와 f5가 20 이상인 f1의 평균
mean_f1 = age_top20[(age_top20['f4'] == 'ISFJ') & (age_top20['f5'] >= 20)]['f1'].mean()
mean_f1
```
