주어진 데이터 셋에서 f2가 0값인 데이터를 age를 기준으로 오름차순 정렬하고 <br>
앞에서 부터 20개의 데이터를 추출한 후 <br>
f1 결측치(최소값)를 채우기 전과 후의 분산 차이를 계산하시오 (소수점 둘째 자리까지) <br>

- 데이터셋 : basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작
- File -> Editor Type -> Script


```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("basic1.csv")

df.head()
```

```python
df_f0 = df[df['f2'] == 0]
df_f0.head()
```

```python
df_age = df_f0.sort_values('age').head(20)
df_age
```

```python
df_min = df_age['f1'].min()
df_min
```

```python
df_fill = df_age['f1'].fillna(df_min)
df_fill
```

```python
fill_res = df_fill.var()
age_res = df_age['f1'].var()

print(fill_res, age_res)
print(round(age_res - fill_res , 2))
```
