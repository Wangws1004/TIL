```python
# 주어진 데이터의 첫 행부터 순서대로 80%까지의 데이터를 훈련데이터로 추출 후 'total_bed-rooms' 변수의 결측값을
# 중앙값으로 대체하고 대체 전의 'total_bedrooms'변수 표준편차 값과 대체 후의 'total_bedrooms' 변수 표준편차
# 값의 차이의 절댓값을 구하시오
```

```python
import pandas as pd

df = pd.read_csv('1_2_housing.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df_80 = df[ : int(len(df)*0.8)]
print(len(df_80))
```

```python
df_80.isnull().sum()
```

```python
df_80_bed = df_80['total_bedrooms'].std()
print(df_80_bed)
```

```python
df_80_med = df_80['total_bedrooms'].median()
print(df_80_med)
```

```python
df_80_fill = df_80['total_bedrooms'].fillna(df_80_med)
print(df_80_fill)
```

```python
df_80_bed_after = df_80_fill.std()
print(df_80_bed_after)
```

```python
res = abs(df_80_bed_after - df_80_bed)
res
```

```python
# 인구 컬럼인 population 항목의 이상값의 합계를 계산하시오(이상값은 평균에서 표준편차* 1.5를 초과하거나 미만인값)
```

```python
import pandas as pd

df = pd.read_csv('1_3_housing.csv')
df.head()
```

```python
df_pop = df['population']
pop_mean = df_pop.mean()
pop_std = df_pop.std()

print(pop_mean)
print(pop_std)
```

```python
top = pop_mean + (pop_std * 1.5)
btm = pop_mean - (pop_std * 1.5)

print(top)
print(btm)
```

```python
df_pop_outlier = df_pop[(df_pop > top) | (df_pop < btm)]
print(df_pop_outlier)
```
