```python
# 소방서별 신고접수시간, 출발시간, 도착시간
# 신고접수 후 출발시간까지를 '대응시간'
# 출발시간부터 도착시간까지를 '출동시간'
# 소방서별 출동시간 평균이 가장 큰 소방서를 구하시오
```

```python
import pandas as pd

df = pd.read_csv("6_1_1_Fire Station Data.csv", encoding="CP949")
df.head()
```

```python
df.info()
```

```python
df['신고접수시간'] = pd.to_datetime(df['신고접수시간'])
df['출발시간'] = pd.to_datetime(df['출발시간'])
df['도착시간'] = pd.to_datetime(df['도착시간'])
df.info()
```

```python
# 초로 표현한것을 60으로 나눠서 분으로 변환
df['대응시간'] = (df['출발시간'] - df['신고접수시간']).dt.total_seconds()/60
df['출동시간'] = (df['도착시간'] - df['출발시간'] ).dt.total_seconds()/60
df.head()
```

```python
df_response = df.groupby('소방서ID')[['출동시간']].mean()
df_response = df_response.reset_index() # 인덱스 정리
df_response.head()
```

```python
df_response = df_response.sort_values('출동시간', ascending=False)
df_response.head()
```

```python
res = df_response.iloc[0]['소방서ID']
res
```

```python
import pandas as pd
df2 = pd.read_csv('6_1_3_Crime Data.csv')
df2.head()
```

```python
df2.info()
```

```python
df2['Year'] = pd.to_datetime(df2['Year-Month']).dt.year
df2['Year'].head()
```

```python
df2.loc[:, 'Total'] = df2.loc[:, 'Violent':'Other'].sum(axis=1)
df2
```

```python
crime_sum = df2[['Police Station', 'Year', 'Traffic', 'Total']].groupby(by=['Police Station', 'Year']).sum(['Traffic', 'Total']).reset_index()
crime_sum.head()
```

```python
crime_2021 = df2[df2['Year'] == 2021][['Police Station', 'Year', 'Total']]
crime_2021.head()
```

```python
crime_2022 = df2[df2['Year'] == 2022][['Police Station', 'Year', 'Total']]
crime_2022.head()
```

```python
merged_crime = pd.merge(crime_2021, crime_2022, on='Police Station')
merged_crime.head()
```

```python
merged_crime['Increase'] = merged_crime['Total_y'] - merged_crime['Total_x']
highest_crime = merged_crime[merged_crime['Increase'] == max(merged_crime['Increase'])]['Police Station'].item()
print(highest_crime)
```

```python
res = df2[(df2['Police Station'] == highest_crime) & (df2['Year'] == 2021)]['Traffic'].sum()
print(res)
```
