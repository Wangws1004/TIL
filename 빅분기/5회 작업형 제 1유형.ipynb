```python
# trash bag 데이터 세트 : 지역별 종량제 봉투 가격
# 가격 컬럼은 각 행의 조건을 만족하는 해당 용량의 종량제 봉투가 존재하면 가격을 값으로, 존재하지 않으면 0을 값으로 갖는다.
# 이 때 용도가 '음식물쓰레기'이고 사용 대상이 '가정용'인 2L 봉투 가격의 평균을 소수점을 버린 후 정수로 출력하시오.
```

```python
import pandas as pd

df1 = pd.read_csv('5_1_trash_bag.csv', encoding='cp949')
df1.head()
```

```python
filtered_data = df1[(df1['용도'] == '음식물쓰레기') & (df1['사용대상'] == '가정용')]
filtered_data.head()
```

```python
ave = filtered_data[filtered_data['2L가격'] != 0]['2L가격'].mean()
print(ave)
res = int(ave)
print(res)
```

```python
# BMI 지수
# 몸무게(KG)를 키(M)의 제곱으로 나누어 구한다

# 주어진 BMI세트에서 비만도가 정상에 속하는 인원 수와 과체중에 속하는 인원 수의 차이를 정수로 출력

```

```python
import pandas as pd

df2 = pd.read_csv('5_2_bmi.csv')
df2
```

```python
df2.info()
```

```python
df2['BMI'] = df2['Weight'] / ((df2['Height']/100) ** 2)
df2.head()
```

```python
# 정상에 속하는 인원 수와 과체중에 속하는 인원 수

# 18.5 이상 23 미만 = 정상
# 23 이상 25미만 = 과체중

norm = df2[(df2['BMI'] >= 18.5) & (df2['BMI'] < 23)]
over = df2[(df2['BMI'] >= 23) & (df2['BMI'] < 25)]

print(len(norm) - len(over))
```

```python
# students 데이터 세트
# 각 학교의 학년별 총 전입학생, 총 전출학생, 전체 학생 수
# 순 전입학생 수는 총 전입학생 수에서 총 전출학생 수를 빼서 구할수 있다.
# 순 전입학생이 "가장 많은 학교의 전체 학생 수"를 구하시오
```

```python
import pandas as pd

df3 = pd.read_csv('5_3_students.csv', encoding='CP949')
df3.head()
```

```python
df3.info()
```

```python
df3.isnull()
```

```python
df3['순 전입학생'] = df3['총 전입학생'] - df3['총 전출학생']
df3.head()
```

```python
school = df3.groupby('학교').agg({'순 전입학생':'sum', '전체 학생 수':'sum'}).reset_index()
school
```

```python
res = school.loc[school['순 전입학생'].idxmax(), '전체 학생 수']
res
```
