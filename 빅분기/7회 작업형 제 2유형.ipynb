```python
# 아파트 데이터 가격
# 아파트 가격을 결정하는 10개의 독립변수와 1개의 종속변수
# 각 독립변수의 이상치를 사분위수를 기준으로 측정하고
# 이상치가 가장 많은 독립변수를 구하시오
# 이때 결측치를 먼저 제거한 후 이상치를 구하시오
# 이상치 : (1Q - 1.5IQR)보다 작거나, (3Q + 1.5IQR)보다 큰 값
# IQR  : Q3 - Q1
```

```python
import pandas as pd

df3 = pd.read_csv('7_1_3_apartment_prices_dataset.csv')
df3.head()
```

```python
df3.info()
```

```python
df3_drop = df3.dropna()
```

```python
df3_drop.info()
```

```python
# 이상치 : (1Q - 1.5IQR)보다 작거나, (3Q + 1.5IQR)보다 큰 값
# IQR  : Q3 - Q1

outliars_count = {}
for col in df3_drop[:-1]:
  Q1 = df3_drop[col].quantile(0.25)
  Q3 = df3_drop[col].quantile(0.75)
  IQR = Q3 - Q1
  low = Q1 - IQR*(1.5)
  top = Q3 - IQR*(1.5)

  outliar = df3_drop[(df3_drop[col] < low) | (df3_drop[col] > top)]
  outliars_count[col] = outliar.shape[0]
```

```python
res = max(outliars_count, key=outliars_count.get)
print(res)
```

```python
# 기존 영업지점에서 데이터를 이용하여 신규 영업지점을 설립할 경우 달성 가능한 매출을 예측
# 주어진 데이터 세트를 활용해 매출 예측 머신러닝 모델을 구축
# 테스트 데이터를 대상으로 매출을 예측한 결과를 CSV파일로 제출
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_squared_error
```

```python
train_data = pd.read_csv('7_2_1_sales_train_dataset.csv')
test_data_x = pd.read_csv('7_2_2_sales_test_dataset_x.csv')
test_data_y = pd.read_csv('7_2_3_sales_test_dataset_y.csv')
```

```python
train_data.head()
```

```python
# 범주형
categorical = ["BranchName", "City", "IndustryType"]

# 수치형
numeric = ["Population", "IncomeGeneratingPopRatio", "AverageIncome"]
```

```python
# 원-핫 인코딩
combined = pd.concat([train_data, test_data_x])
combined_encoded = pd.get_dummies(combined, columns=categorical)
```

```python
# 인코딩된 데이터 분리
train_data_encoded = combined_encoded.iloc[:len(train_data)]
test_data_x_encoded = combined_encoded.iloc[len(train_data):].drop(['Sales'], axis=1)
```

```python
# 수치형 데이터 스케일링
scaler = StandardScaler()
train_data_encoded[numeric] = scaler.fit_transform(train_data_encoded[numeric])
test_data_x_encoded[numeric] = scaler.transform(test_data_x_encoded[numeric])
```

```python
# 매출 예측
X_train_encoded = train_data_encoded.drop('Sales', axis=1)
y_train = train_data_encoded['Sales']
X_test_encoded = test_data_x_encoded
y_test = test_data_y['Sales']
```

```python
# 선형회귀
linear_model = LinearRegression()
linear_model.fit(X_train_encoded, y_train)
linear_pred = linear_model.predict(X_test_encoded)
linear_rmse = np.sqrt(mean_squared_error(y_test, linear_pred))

print("Linear Regression RMSE:", linear_rmse)
```

```python
# 랜덤포레스트
random_forest_model = RandomForestRegressor(n_estimators=100)
random_forest_model.fit(X_train_encoded, y_train)
rf_pred = random_forest_model.predict(X_test_encoded)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
print("Random Forest RMSE:", rf_rmse)
```

```python
# MLP
mlp_model = MLPRegressor(hidden_layer_sizes=(50,), max_iter=500)
mlp_model.fit(X_train_encoded, y_train)
mlp_pred = mlp_model.predict(X_test_encoded)
mlp_rmse = np.sqrt(mean_squared_error(y_test, mlp_pred))
print("MLP Regressor RMSE:", mlp_rmse)
```

```python
res = random_forest_model.predict(X_test_encoded)

result_df = pd.DataFrame({'PredictedSales': res})
result_df.to_csv('000.csv', index=False)
```
