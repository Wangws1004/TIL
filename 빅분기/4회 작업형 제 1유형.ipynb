```python
# CaliforniaHousuing 데이터 세트
# 1. 데이터 중 결측치가 있는 경우 데이터의 행을 모두 제거하고
# 2. 첫 번째 행부터 순서대로 70%까지의 데이터를 훈련 데이터로 추출
# 3. 'housing_median_age'의 제1사분위수 값을 정수로 계산
```

```python
import pandas as pd

df = pd.read_csv('2_1_housing.csv')
df.head()
```

```python
df.info()

# total_bedrooms -> 결측치 존재
# ocean_proximity -> 데이터 타입 object
```

```python
df_dropna = df.dropna()
print(len(df))
print(len(df_dropna))
```

```python
# 첫 번째 행부터 순서대로 70% 추출
df_dropna_70 = df_dropna.iloc[:int(len(df_dropna)*0.7)]
print(len(df_dropna_70))
```

```python
# 'housing_median_age'의 제1사분위수 값을 정수로 계산
res = int((df_dropna_70['housing_median_age'].quantile(0.25)))
print(res)
```

```python
# 국가별 연도별 인구 10만 명당 결핵 유병률 데이터 세트
# 2000년도의 국가별 결핵 유병률 데이터 세트에서 2000년도의 평균값보다 더 큰 유병률을 가진 국가의 수를 계산
```

```python
import pandas as pd

df2 = pd.read_csv('2_2_worlddata.csv')
df2.head()
```

```python
# 2000년도의 데이터 추출
df_year_2000 = df2[df2['year'] == 2000]
df_year_2000
```

```python
df_year_2000_T = df_year_2000.T
df_year_2000_T
```

```python
df_year_2000_T = df_year_2000_T.drop(['year'])
print(df_year_2000_T)
```

```python
df_year_2000_T.describe()
```

```python
df_year_2000_T.columns = ['value']
df_year_2000_T
```

```python
mean = df_year_2000_T['value'].mean()
```

```python
res = len(df_year_2000_T[df_year_2000_T['value']>mean])
print(res)
```
